```python
import matplotlib.pyplot as plt

def plot_pie_chart(data, labels, title="Pie Chart"):
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title(title)
    plt.show()
    
Material_grade = [10, 20, 30, 40]
labels = ['Composites', 'Steel', 'Polymers', 'Glass']
plot_pie_chart(Material_grade, labels, title="Material Grade Chart")
```
